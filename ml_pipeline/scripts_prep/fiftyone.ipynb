```python
import fiftyone as fo

# Check if it exists, then load it; otherwise, create it
if fo.dataset_exists("cat-playground-dataset"):
    dataset = fo.load_dataset("cat-playground-dataset")
else:
    dataset = fo.Dataset.from_dir(
        dataset_dir="dataset/crop",
        dataset_type=fo.types.ImageDirectory,
        name="cat-playground-dataset"
    )

session = fo.launch_app(dataset, auto=False)
print(session.url)
```
